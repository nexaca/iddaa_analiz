# ALT UST & ORAN ILISKISI 

Soru : alt ust oranlari ile alt ust arasnda bir iliski var m ? 

```python
alt_ust_df = df[['Div',
 'Date',
 'Time',
 'HomeTeam',
 'AwayTeam',
 'FTHG',
 'FTAG',
 'FTR',
 'HTHG',
 'HTAG',
 'HTR',
 'B365H',
 'B365D',
 'B365A',
 'B365>2.5',
 'B365<2.5']]
```

```python
alt_ust_df['TG'] = alt_ust_df["FTHG"]+ alt_ust_df["FTAG"]
```

```python
def alt_ust(skor):
    if skor > 2.5:
        return 'ust'
    else:
        return 'alt'


alt_ust_df['alt_ust'] = alt_ust_df['TG'].apply(alt_ust)
```

```python
alt_ust_df.head(20)
```

```python
ust_mac = alt_ust_df['alt_ust'] == 'ust'
```

```python
alt_ust_df.groupby(by = ["B365<2.5", "B365>2.5", 'alt_ust','B365H','B365D','B365A']).count()
```

```python
alt_oran  = alt_ust_df['B365>2.5'] == 1.75
ust_oran  = alt_ust_df['B365<2.5'] == 2.05

alt_ust_df[alt_oran & ust_oran].sort_values(by ='alt_ust')
```
